THE FOLLOWING LINES ARE ALL STANDARD

```python
detector = ObjectDetection()
```

```python
detector.setModelTypeAsYOLOv3()
```

```python
detector.setModelPath("/content/drive/MyDrive/colab/yolo.h5")
```

```python
detections = detector.detectObjectsFromImage(input_image="/content/drive/MyDrive/colab/img7.jpg", output_image_path="imagenew7.jpg", minimum_percentage_probability=30)
```

```python
detector.loadModel()
```
