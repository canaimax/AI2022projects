```python
from imageai.Detection import ObjectDetection

detector = ObjectDetection()
```

```python
detector.setModelTypeAsYOLOv3()
```

```python
detector.setModelPath("/content/drive/MyDrive/colab/yolo.h5")
```

```python
detector.loadModel()
```

```python
detections = detector.detectObjectsFromImage(input_image="/content/drive/MyDrive/colab/img8.jpg", output_image_path="imagenew.jpg", minimum_percentage_probability=30)
```
